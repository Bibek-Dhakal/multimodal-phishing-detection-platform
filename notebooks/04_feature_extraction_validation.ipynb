# Offline Feature Extraction Validation

In this notebook, we instantiate our custom `FeatureExtractor` class and validate that it correctly transforms raw URL strings into the exact **79-dimensional mathematical matrix** required by our ISCX-trained XGBoost model.

We will test it on a Legitimate URL and a Phishing URL to ensure no `NaN` or `Infinity` errors occur during safe division.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))

from src.feature_extractor import FeatureExtractor
```

### Test 1: Benign / Whitelisted URL
Top-tier enterprise domains should immediately return a safe array of zeros, protecting the system against False Positives.

```python
safe_url = "https://www.google.com/search?q=machine+learning"
extractor_safe = FeatureExtractor(safe_url)
vector_safe = extractor_safe.extract()

print(f"Generated Vector Shape: {vector_safe.shape}")
print(f"Whitelist Status: {extractor_safe.is_whitelisted()}")
print("\nVector Output:")
print(vector_safe)
```

### Test 2: Unknown Malicious Typosquatting URL
An unknown, un-whitelisted URL must be parsed through all 79 rules, catching lengths, entropies, digit transitions, and query ratios.

```python
malicious_url = "http://secure-update-paypa1.com.biz/login.php?session=893247982374&verify=true"
extractor_mal = FeatureExtractor(malicious_url)
vector_mal = extractor_mal.extract()

print(f"Generated Vector Shape: {vector_mal.shape}")
print(f"Whitelist Status: {extractor_mal.is_whitelisted()}")
print("\nFeature Matrix Output (79-Dimensions):")
for i, val in enumerate(vector_mal):
    print(f"Feature {i+1:02d}: {val:.4f}")
```
